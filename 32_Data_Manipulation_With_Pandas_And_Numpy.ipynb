```python
## renaming columns

df=df.rename(columns={7:8})
df.head()
```

```python
## data aggregating and grouping
df.head()
```

```python
grouped_mean=df.groupby('7')['0.4'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['7','0.4'])['0.2'].sum()
print(grouped_sum)
```

```python
grouped_sum=df.groupby(['7','0.4'])['0.2'].mean()
print(grouped_sum)
```
