# Pandas-dataframe and series
Pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. it provides two primary data structures : series and data frame . a series is a one-dimensional array-like object, while a dataframe is a two-dimensional,size-mutable, and potentially heterogenous tabular data structure with labeled axes (rows and columns)

```python
import pandas as pd
```

```python
## Series
## A pandas series is a one-dimensional array-like object that can hold any data type. it is similar to a column in a table

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
## create a series from dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## dataframe
## create a dataframe from a dictionary of list

data={
    'name':["chetan","gaurav","dhananjay"],
    'age':[20,21,20],
    'city':["pune","pune","nagpur"]
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## create a dataframe from a list of dict

data=[
    {'name':"chetan","age":20,"city":"mumbai"},
    {'name':"dhananjay","age":20,"city":"delhi"},
    {'name':"gaurav","age":20,"city":"mumbai"},
    {'name':"archana","age":20,"city":"pune"},
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
